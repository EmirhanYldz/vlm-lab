```python
import wikipedia

def search_wikipedia(query, num_sentences=3):
    try:
        summary = wikipedia.summary(query, sentences=num_sentences)
        return summary
    except wikipedia.exceptions.DisambiguationError as e:
        return wikipedia.summary(e.options[0], sentences=num_sentences)
    except Exception as e:
        return "No relevant information found."

query = "Cats play behavior"
info = search_wikipedia(query)
print(info)
```

```python
import wikipedia

def get_wikipedia_pages(topic, max_pages=5):
    related_pages = []
    try:
        page = wikipedia.page(topic)
        related_pages.append(page.content)
        links = page.links

        for link in links[:max_pages - 1]:
            try:
                sub_page = wikipedia.page(link)
                related_pages.append(sub_page.content)
            except:
                continue

        return related_pages
    except Exception as e:
        print(f"Error retrieving Wikipedia pages: {e}")
        return related_pages

pages = get_wikipedia_pages("Cat weight")
```

```python
import wikipedia
import mwparserfromhell
from bs4 import BeautifulSoup

def parse_wikipedia_page(title):
    try:
        page = wikipedia.page(title)
        raw_content = page.content

        parsed_content = mwparserfromhell.parse(raw_content)
        clean_content = parsed_content.strip_code()  

        soup = BeautifulSoup(clean_content, 'html.parser')
        text_content = soup.get_text(separator="\n")

        return {
            "title": page.title,
            "content": text_content,
            "url": page.url
        }
    except Exception as e:
        print(f"Error parsing page {title}: {e}")
        return None

titles = wikipedia.search("Polar bear, weight", results=10)
parsed_pages = [parse_wikipedia_page(title) for title in titles if parse_wikipedia_page(title) is not None]
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')

def embed_page_content(page_content):
    return model.encode(page_content, convert_to_tensor=True)

embeddings = [embed_page_content(page) for page in pages]
```

```python
%pip install faiss-cpu
```

```python
import faiss
import numpy as np

embedding_dim = embeddings[0].shape[0]

index = faiss.IndexFlatL2(embedding_dim)

embeddings_array = np.array([embedding.cpu().numpy() for embedding in embeddings])
index.add(embeddings_array)
```

```python
def retrieve_relevant_info(user_query, top_k=3):
    query_embedding = model.encode(user_query, convert_to_tensor=True).cpu().numpy()

    _, indices = index.search(np.array([query_embedding]), top_k)

    relevant_pages = [pages[idx] for idx in indices[0]]
    return relevant_pages

query = "How does machine learning work?"
relevant_info = retrieve_relevant_info(query)
print(relevant_info)
```

```python
def prepare_input_for_model(visual_description, user_question, retrieved_info):
    context = "\n\n".join(retrieved_info)
    input_text = f"{visual_description}\n\nUser Question: {user_question}\n\nContext: {context}"
    return input_text

input_for_model = prepare_input_for_model("A picture of a robot.", query, relevant_info)
```
